```python
recomb_file = Path("../data/recombinants.csv")
recomb_df = pd.read_csv(recomb_file)
recomb_df
```

```python
#import tskit
#ts_file = Path("../data/v1-beta1_2023-02-21.ts")
#ts = tskit.load(ts_file)
import tszip
ts_file = Path("../data/sc2ts_viridian_v1.2.trees.tsz")
ts = tszip.decompress(ts_file)
ts
```

```python
site_map = {int(x): y for x, y in zip(ts.sites_position, range(ts.num_sites))}
#site_map
```

```python
focal_node_ids = np.sort(np.unique(
    recomb_df.recombinant.to_list() +\
    recomb_df.parent_left.to_list() +\
    recomb_df.parent_right.to_list()
))
sts, node_map = ts.simplify(focal_node_ids, filter_sites=False, map_nodes=True)
sts
```

```python
data = []
for _, row in tqdm.tqdm(recomb_df.iterrows(), total=len(recomb_df)):
    re_node = node_map[row.recombinant]
    parent_left = node_map[row.parent_left]
    parent_right = node_map[row.parent_right]
    if row.interval_left == 27577:
        # TODO: Investigate why this site position is not in ts.
        interval_lft_remap = site_map[row.interval_left - 1]
    else:
        interval_lft_remap = site_map[row.interval_left]
    interval_rgt_remap = site_map[row.interval_right]
    trio_seqs = sts.genotype_matrix(
        samples=[re_node, parent_left, parent_right],
        alleles=("A", "C", "G", "T", "-"),
    )
    mns_left = []
    mns_right = []
    for x in mns:
        is_in_lft = 0
        is_in_rgt = 0
        x_len = len(x["mutations"])
        for m in x["mutations"]:
            pos = m["position"]
            if pos in site_map:
                derived_allele = "ACGT".index(m["derived"])
                pos_remap = site_map[pos]
                if (trio_seqs[pos_remap, 0] == derived_allele) and\
                    (pos <= interval_lft_remap) and\
                    (trio_seqs[pos_remap, 0] == trio_seqs[pos_remap, 1]) and\
                    (trio_seqs[pos_remap, 0] != trio_seqs[pos_remap, 2]):
                    is_in_lft += 1
                if (trio_seqs[pos_remap, 0] == derived_allele) and\
                    (pos >= interval_rgt_remap) and\
                    (trio_seqs[pos_remap, 0] == trio_seqs[pos_remap, 2]) and\
                    (trio_seqs[pos_remap, 0] != trio_seqs[pos_remap, 1]):
                    is_in_rgt += 1
        if is_in_lft == x_len:
            mns_left.append(x["id"])
        if is_in_rgt == x_len:
            mns_right.append(x["id"])
    num_mns_left = len(mns_left)
    num_mns_right = len(mns_right)
    datum = {
        "recombinant": row.recombinant,
        "num_mns_left": num_mns_left,
        "mns_left": ",".join(mns_left),
        "num_mns_right": num_mns_right,
        "mns_right": ",".join(mns_right)
    }
    data.append(datum)
data = pd.DataFrame(data)
data
```

```python
merged_df = pd.merge(recomb_df, data, on='recombinant', how='inner')
merged_df
```

```python
pd.set_option('display.max_rows', 1000)
merged_df[
    [
        'recombinant',
        'net_min_supporting_loci_lft_rgt_ge_4',
        'net_min_supporting_loci_lft',
        'net_min_supporting_loci_rgt',
        'num_mns_left',
        'mns_left',
        'num_mns_right',
        'mns_right'
    ]
].sort_values(
    by=['net_min_supporting_loci_lft_rgt_ge_4', 'num_mns_right', 'num_mns_left'],
    ascending=False
).reset_index()
```

```python
num_good_recomb_total = np.sum(merged_df.net_min_supporting_loci_lft_rgt_ge_4)
num_good_recomb_with_mns = np.sum(
    np.logical_and(
        merged_df.net_min_supporting_loci_lft_rgt_ge_4,
        np.logical_or(merged_df.num_mns_left > 0, merged_df.num_mns_right > 0)
    )
)
perc_good_recomb_with_mns = num_good_recomb_with_mns / num_good_recomb_total
num_bad_recomb_total = np.sum(~merged_df.net_min_supporting_loci_lft_rgt_ge_4)
num_bad_recomb_with_mns = np.sum(
    np.logical_and(
        ~merged_df.net_min_supporting_loci_lft_rgt_ge_4,
        np.logical_or(merged_df.num_mns_left > 0, merged_df.num_mns_right > 0)
    )
)
perc_bad_recomb_with_mns = num_bad_recomb_with_mns / num_bad_recomb_total
print(f"Good recomb. with MNS: {num_good_recomb_with_mns} / {num_good_recomb_total} ({perc_good_recomb_with_mns}%)")
print(f" Bad recomb. with MNS: {num_bad_recomb_with_mns} / {num_bad_recomb_total} ({perc_bad_recomb_with_mns}%)")
```
