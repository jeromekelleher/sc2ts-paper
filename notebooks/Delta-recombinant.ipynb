# Explore the recombinant origins of B.1.617.2

````python
import numpy as np
import tskit
import tszip
import pandas as pd

import matplotlib.pyplot as plt
import sys

sys.path.append("../")
import sc2ts
````

````python
ts_path  ="../results/DELETE_v1-alpha4-mds1000/DELETE_v1-alpha4-mds1000_2021-06-21.ts.tsz"
ds_path = "../data/viridian_2024-04-29.alpha1.zarr.zip"
ts = tszip.load(ts_path)
ts
````

````python
ti = sc2ts.TreeInfo(ts)

def report(*args, **kwargs):
    display(*ti.node_report(*args, **kwargs))

````

````python
first_617_2 = ti.pango_lineage_samples["B.1.617.2"][0]
focal_strain = ti.nodes_metadata[first_617_2]["strain"]
````

````python
report(first_617_2)
````

````python
%%time
ts_path  ="../results/DELETE_v1-alpha4-mds1000/DELETE_v1-alpha4-mds1000_2021-03-22.ts"

import itertools
runs = []
for direction, num_mismatches in itertools.product(["forward", "reverse"], [3, 4, 5, 6, 7, 8, 9]):
    run = sc2ts.run_hmm(ds_path, ts_path, [focal_strain], direction=direction, num_mismatches=num_mismatches, 
                       deletions_as_missing=True, mismatch_threshold=40)
    runs.append(run)

len(runs)
````

````python
data = []
for run in runs:
    r = run[0]
    data.append({
        "direction": r.direction,
        "num_mismatches": r.num_mismatches,
        "parents": len(r.match.path),
        "mutations": len(r.match.mutations)
    })
df = pd.DataFrame(data)
df
````

For num_mismatches of < 6 we get a recombinant with 24 mutations and for greater >= 6 we have 29 mutations. How stable are these matches?

````python
data = []
for run in runs:
    r = run[0]
    if len(r.match.path) > 1:
        data.append({
            "direction": r.direction,
            "num_mismatches": r.num_mismatches,
            "left_parent": r.match.path[0].parent,
            "right_parent": r.match.path[1].parent,
            "breakpoint": r.match.path[0].right,
            "mutations": [str(m) for m in r.match.mutations]
    })
dfr = pd.DataFrame(data)
dfr
````

We do get a somewhat different solution on the reverse HMM pass, both in terms of breakpoint interval and the parent nodes. The breakpoint interval for the forward pass is 25277-25469, whereas the breakpoint on the reverse run is 22023 (we don't have an interval for that, currently).

````python
data = []
for run in runs:
    r = run[0]
    if len(r.match.path) == 1:
        data.append({
            "direction": r.direction,
            "num_mismatches": r.num_mismatches,
            "parent": r.match.path[0].parent,
            "mutations": [str(m) for m in r.match.mutations]
    })
df_nor = pd.DataFrame(data)
df_nor
````

````python
assert len(set([tuple(l) for l in df_nor.mutations])) == 1
````

The no recombinatin solutions are all the same.

Let's just filter down to the matches for num_mismatches = 5 and 6

````python
hmm_runs = [r[0] for r in runs if r[0].num_mismatches in (5, 6)]
assert len(hmm_runs) == 4
````

````python
fwd_re = hmm_runs[0]
rev_re = hmm_runs[2]
no_re = hmm_runs[1]
assert len(fwd_re.match.path) == 2
assert len(rev_re.match.path) == 2
assert len(no_re.match.path) == 1
````

````python
fwd_re_mutations = set([str(m) for m in fwd_re.match.mutations])
rev_re_mutations = set([str(m) for m in rev_re.match.mutations])
no_re_mutations = set([str(m) for m in no_re.match.mutations])
````

````python
common = sorted(fwd_re_mutations & rev_re_mutations & no_re_mutations, key=lambda m: int(m[1:-1]))
common, len(common)
````

There are 19 mutations common to all three solutions.

````python
fwd_re_mutations - rev_re_mutations
````

````python
rev_re_mutations - fwd_re_mutations
````

````python
no_re_mutations - (fwd_re_mutations | rev_re_mutations)
````
